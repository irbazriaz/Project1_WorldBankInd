```python
import pandas as pd
%matplotlib inline
df = pd.read_csv('Data Files/urban-and-rural-population.csv')
```

```python
df
```

```python
Number_of_Years = list(df['Year'].unique())
```

```python
rural_population = []
urban_population = []
for i in Number_of_Years:
    grp = df[df['Year']==i]
    rural_population.append(grp['Rural population'].sum())
    urban_population.append(grp['Urban population'].sum())
```

```python
output_notebook()
```

```python
df1 = pd.DataFrame(Number_of_Years)
df1['Rural Population'] = pd.DataFrame(rural_population)
df1['Urban Population'] = pd.DataFrame(urban_population)
df1 = df1.rename(columns={'0':'Years'})
df1=df1.rename(columns={0:'Years'})
```

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource, CustomJS,Slider
from bokeh.layouts import column
from bokeh.models import HoverTool
from bokeh.models import LinearAxis, Range1d
#source = ColumnDataSource(df1)

source = ColumnDataSource(
        data=dict(
            x=df1['Years'],
            y=df1['Rural Population'],
            z=df1['Urban Population'],
        )
    )


hover = HoverTool(
        tooltips=[
            ("Years:","@x"),
            (".Rural Population:","@y{0.0}"),
            (".Urban Population:","@z{0.0}")

        ]
       # mode='vline'
    )

# create a blank figure
p = figure(plot_width = 600, plot_height=300,tools=[hover],toolbar_location=None, title='Hovering over point:')

# add a line rendrer with x, y coordinates
p.line('x','z',line_color = 'red',line_width=2,alpha=0.4,legend_label='Urban Population',source=source)
p.line('x','y',line_color = 'blue',line_width=2, alpha=0.4, legend_label='Rural Population',source=source)

p.circle('x','z',fill_color='red',size=5,source=source)
p.circle('x','y',fill_color='blue',size=5,source=source)

#slider = Slider(start=1960, end=2020, value=1, step=1, title="foo")


#update_curve = CustomJS(args=dict(source=source, slider=slider), code="""
#    var data = source.data;
#    var f = slider.value;
#    var x = data['x']
#    var y = data['z']
 #   for (var i = 0; i < x.length; i++) {
           
  #  }
    
   # // necessary becasue we mutated source.data in-place
    #source.change.emit();
#""")


p.xaxis.axis_label = "Years"
#p.yaxis.visible=False
#p.yaxis.ticker=[500.0 ,1 ,1.5 ,2.0 ,2.5 ,3.0 ,3.5 ,4.0 ]
p.axis.major_label_text_color = "green"
p.extra_y_ranges = {"foo": Range1d(start=500, end= 4500)}
p.add_layout(LinearAxis(y_range_name="foo"), 'left')
p.legend.location = 'top_left'
p.legend.label_text_font = "times"
p.legend.label_text_font_style = "italic"
p.legend.label_text_color = "navy"



show(p)
```
