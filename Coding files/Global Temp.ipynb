```python
# Import required libs
import pandas as pd
```

```python
datafile = 'Data Files/data.json'
```

```python
df = pd.read_json(datafile,orient='records')
```

```python
df
```

### Data Cleansing Part

```python
# Remove unwanted column
del df['description']
```

```python
#Drop if there are any missing values
df = df.dropna(how='any')
```

```python
df.set_index=None
```

```python
x = df.index
x
y=df['data']
y
```

```python
# Import Bokeh Libraries
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
output_notebook()
from bokeh.models import ColumnDataSource

from bokeh.models import HoverTool

source = ColumnDataSource(
        data=dict(
            x=df.index,
            y=df['data']
        )
    )

hover = HoverTool(
        tooltips=[
            
            ("Temp:", "$y")
        ]
    )

p = figure(plot_width=800, plot_height=400, tools=[hover], title="Global Land and Ocean Temperatures Anomalies")

p.line('x', 'y', line_width=2, source=source,legend_label='Years',line_color='red')
p.circle('x','y',fill_color='white',size=8,source=source)


show(p)
```

```python
import numpy as np
from bokeh.models.annotations import Span
from math import pi

data = list((df.index))
y = df['data']

from bokeh.models import HoverTool

x = df.index
y = df['data']

hover = HoverTool(
        tooltips=[
            
            ("Temp:", "$y")
        ]
    )

# Set the x_range to the list of categories above
p = figure(y_range=(-1,1),x_range=(data), plot_height=250,plot_width=800,tools=[hover], title="Global Land and Ocean Temperatures Anomalies")

# Categorical values can also be used as coordinates
p.vbar(x=(data), top=y, width=0.2,)

lower = Span(location=0, dimension='width', line_color='firebrick', line_width=4)
p.add_layout(lower)

p.xaxis.major_label_orientation = pi/4
# Set some properties to make the plot look better
p.xgrid.grid_line_color = None
#p.y_range.start = 0

show(p)
```
