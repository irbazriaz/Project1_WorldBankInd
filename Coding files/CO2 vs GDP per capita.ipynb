```python
import pandas as pd
import humanize
```

```python
datafile = 'Data Files/co2-emissions-vs-gdp.csv'
```

```python
pd.set_option('display.max_rows',10)
df = pd.read_csv(datafile)
df
```

```python
df_2016 = df[df['Year']==2016]
humanize.intword(df['Total population (Gapminder, HYDE & UN)'])
#df_2016 = humanize.intword(df['Total population (Gapminder, HYDE & UN)'])
df_2016
```

```python
total_ppopulation = []
for value in df_2016['Total population (Gapminder, HYDE & UN)'].iteritems():
    total_ppopulation.append(humanize.intword(value))

total_ppopulation
```

```python
df1 = pd.DataFrame(df_2016['Entity'])
df1['Total Population'] = pd.DataFrame(df['Total population (Gapminder, HYDE & UN)'])
df1['GDP Per Capita'] = pd.DataFrame(df['Real GDP per capita in 2011US$, 2011 benchmark (Maddison Project Database (2018))'])
df1['CO2 emissions Per capita '] = pd.DataFrame(df['Per capita CO2 emissions'])
```

```python
df1 = df1.rename(columns={'Total Population':'Total_Population',
                         'GDP Per Capita':'GDP_Per_Capita',
                         'CO2 emissions Per capita':'CO2'})
```

```python
df1 = df1.dropna(how='any')
df1
```

```python
from bokeh.sampledata.autompg import autompg
from bokeh.models import LinearColorMapper, ColorBar,ColumnDataSource
from bokeh.transform import transform
from bokeh.plotting import figure,show,output_notebook
from bokeh.models import LinearAxis, Range1d

output_notebook()

source = ColumnDataSource(df1)
color_mapper = LinearColorMapper(palette="Viridis256", low=df1['Total_Population'].min(), high=df1['Total_Population'].max())

p = figure(x_axis_label='GDP Per Capita', y_axis_label='CO2 emissions Per capita', tools='', toolbar_location=None)
p.circle(x='GDP_Per_Capita', y='CO2 emissions Per capita ', color=transform('Total_Population', color_mapper), size=20, alpha=0.6, source=df1)

color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0,0), title='Total Population')
p.add_layout(color_bar, 'right')

show(p)
```

```python

import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(16,8))
plt.scatter(df1['GDP_Per_Capita'],df1['CO2 emissions Per capita '],cmap='viridis',facecolors="red", edgecolors="black")
plt.style.use('seaborn-darkgrid')
plt.xlabel('GDP Growth Per Capita')
plt.ylabel('CO2 emissions per capita')
```

```python
import numpy as np
import scipy.stats as st
# Demonstrate calculating the variance and standard deviation using the different modules
var_numpy = np.var(df1['GDP_Per_Capita'],ddof = 0)
print(f"The GDP Per capita variance using the NumPy module is {var_numpy}")

sd_numpy = np.std(df1['CO2 emissions Per capita '],ddof = 0)
print(f"The CO2 emissions Per capita standard deviation using the NumPy module is {sd_numpy}")
```

```python
# The next example will compute the Pearson correlation coefficient between "GDP per Capita" and "C02 Emission per capita"
gdp_per_capita = df1.iloc[:,2]
co2_emission_per_capita = df1.iloc[:,3]
correlation = st.pearsonr(gdp_per_capita,co2_emission_per_capita)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
from scipy.stats import linregress
x_values = df1['GDP_Per_Capita']
y_values = df1['CO2 emissions Per capita ']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(0,50),fontsize=15,color="red")
plt.xlabel('GDP_Per_Capita')
plt.ylabel('CO2 emissions Per capita')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
